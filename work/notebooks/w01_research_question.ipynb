## 2. The question: decision, action, cost of a wrong call

**1. What decision does this work improve?**
It improves the decision: *"Which specific page should our editorial team review and update this week to reverse traffic decline?"* Instead of choosing pages randomly, by age, or by simple guesswork, editors can work down a prioritized queue of pages ranked by their recovery potential and risk of decline.

**2. Who acts on the output, and what do they do?**
The primary actors are **Content Editors** and **SEO Managers**. Based on the prioritized list and the associated reason codes (e.g., `ctr_review_candidate`, `stale_visible_page`, `thin_visible_page`), they will audit the target page, update out-of-date information, expand thin content sections, or rewrite titles and meta descriptions.

**3. What does a wrong recommendation cost?**
- **False Positive (recommending a page that doesn't need refresh or has no potential):** Costs wasted editor hours (typically 2–4 hours per page) spent rewriting content that was already stable or had no traffic opportunity.
- **False Negative (missing a page that is heavily declining):** Costs lost traffic, clicks, conversions, and revenue as high-visibility pages degrade in ranking without intervention.

**4. Why does data/ML help?**
A simple rule (e.g. `days_since_last_update > 180`) is insufficient. Content decay is a multi-dimensional problem where signals like average position, CTR tiers, word counts, and search volume interact in non-linear ways. ML helps by learning these patterns across thousands of pages to output a highly accurate risk score.

## 3. Quick look at the data (2-3 real numbers)

Let's load the starter dataset `data/raw/content_refresh_anonymized.csv` to compute 2-3 real numbers showing why this lane is worth our focus over the next 7 weeks. We look at the scale of decline, the volume risk, and the typical age of declining content.

```python
import pandas as pd
import numpy as np

# Load starter data (note we are running in work/notebooks/ so path is relative to that)
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Define target label for decline
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

# 1. Total and declining rate
total_pages = len(df)
declining_pages = df["is_declining_label"].sum()
decline_rate = df["is_declining_label"].mean()

print(f"Total pages in starter dataset: {total_pages:,}")
print(f"Declining pages: {declining_pages:,} ({decline_rate:.1%})")

# 2. Declining pages that are highly visible
visible_declining = df[(df["is_declining_label"] == 1) & (df["impressions_90d"] >= 500)]
num_visible_declining = len(visible_declining)
pct_visible_declining = num_visible_declining / declining_pages

print(f"Declining pages with >= 500 impressions (high visibility): {num_visible_declining:,} ({pct_visible_declining:.1%} of declining pages)")

# 3. Median and average age of declining visible pages
avg_age_days = visible_declining["content_age_days"].mean()
median_age_days = visible_declining["content_age_days"].median()

print(f"Average age of declining visible pages: {avg_age_days:.1f} days")
print(f"Median age of declining visible pages: {median_age_days:.1f} days")
```

## 4. Careful words: what I can and can't claim

We must maintain scientific honesty and write using careful language:

- **What we CAN claim**: 
  - *"We observed a strong association between content age, position tier, and traffic decline in our dataset."*
  - *"Our model provides decision-support by ranking pages based on their observed probability of traffic decline."*
  - *"We measured a directional improvement in Precision@50 using a random forest classifier over our rule-based baseline."*
- **What we CANNOT claim**:

  - *"Refreshing this page will cause its traffic to increase."* (Our model ranks pages by risk of decline; it cannot prove causal recovery without an A/B test or randomized experiment.)
  - *"We can predict exactly how many clicks a page will get next month."* (We predict probabilities and ranking tiers, not deterministic future metrics.)
